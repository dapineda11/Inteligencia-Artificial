###**Cálculo de R cuadrado**


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=np.array([1, 1.5, 2, 2.5, 3, 3.5 , 4, 4.5, 5, 5.5, 6])
x0=np.ones(x1.shape)
y=np.array([0.16,0.28,0.38,0.47,0.60,0.73,0.79,0.87,0.84,0.86,0.96])
X=np.matrix([x0,x1]).T
Y=np.matrix([y]).T
Theta=np.linalg.inv(X.T*X)*(X.T)*Y
```

```python
plt.plot(x1,y, 'bo')
plt.plot(x1,Theta[0,0]+Theta[1,0]*x1)
plt.title("Parte real de la impedancia(Z)")
plt.xlabel('Voltaje')
plt.ylabel('Corriente')
plt.show()
```

```python
from sklearn.metrics import r2_score
print("El valor de R^2 es: ",r2_score(y,Theta[0,0]+Theta[1,0]*x1))
```
